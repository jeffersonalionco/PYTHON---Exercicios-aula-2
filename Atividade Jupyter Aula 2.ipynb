```python
### Calculando a média de idade por classe

dfTitanic.groupby('Pclass').mean('Age')
```

```python
### Ver quantos passageiros tem por classe

classeA = dfTitanic[dfTitanic['Pclass'] == 1].size
classeB = dfTitanic[dfTitanic['Pclass'] == 2].size
classeC = dfTitanic[dfTitanic['Pclass'] == 3].size
print('Pclass 1: ', classeA, 'Pclass 2: ', classeB, 'Pclass 3:', classeC)
```

```python
### Somando 10 a cada idade do titanic e criando uma coluna com esses valores

dfTitanic['Idade_10_anos'] = dfTitanic['Age'] + 10
dfTitanic['Idade_10_anos']
```

```python
### Adicionando couna que mostra se o cliente tinha familia

dfTitanic['Familia'] = dfTitanic['SibSp'] > 0
dfTitanic[['Name', 'SibSp', 'Familia']].head()
```

# PARTE 4
### Desafios Extras

```python
Parte 4 - Desafios Extras
Removendo Valores Nulos:
Identifique quais colunas possuem valores nulos.
Substitua os valores nulos da coluna 'Age' pela média das idades.
Ordenação de Dados:
Ordene os passageiros por idade em ordem decrescente.
Ordene por classe e, dentro de cada classe, por idade.
Exportação para Excel:
Salve o DataFrame final em um arquivo Excel chamado titanic_tratado.xlsx.
```

```python
### Removendo valores Nullos
### Isso retorna quantos valor nullo  tem y
dfTitanic.isnull().sum()
```

```python
### Substituir valores null de idade 
dfTitanic['Age'] = dfTitanic['Age'].fillna(dfTitanic['Age'].mean())

dfTitanic
```

```python
### Ordenando os viajantes de forma crescente


dfTitanic.sort_values(by='Age', ascending=True)
```

```python
### Ordenando Viajantes de forma descrecentes
dfTitanic.sort_values(by='Age', ascending=False)

```

```python
### Ordenando por classe e por age

dfTitanic.sort_values(by=['Pclass', 'Age'], ascending=[True, True])
```
